## Task 2)

```python
# Linearly spaced numbers between 0 and 10
t = np.linspace(0,10,51)
```

```python
# Cosines
f = np.cos(t)
```

```python
# Plot the diagram
plt.figure(figsize = (10,5))
plt.plot(t,f, color='green')

plt.title('Graph f(t)', fontsize=16)
plt.xlabel('Values t', fontsize=12)
plt.ylabel('Values y', fontsize=12)

plt.xlim(0.5, 9.5)
plt.ylim(-2.5, 2.5)

plt.show();
```

## Task 3

```python
x = np.linspace(-3,3,51)

y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
# Plot graphs using subplots

fig, ax = plt.subplots(2,2, figsize=(8,6))

ax[0,0].plot(x,y1)
ax[0,1].plot(x,y2)
ax[1,0].plot(x,y3)
ax[1,1].plot(x,y4)

ax[0,0].set_title('Graph y1', fontsize=10)
ax[0,1].set_title('Graph y2', fontsize=10)
ax[1,0].set_title('Graph y3', fontsize=10)
ax[1,1].set_title('Graph y4', fontsize=10)

ax[0,0].set_xlim(-5,5)

plt.subplots_adjust(wspace=0.3, hspace=0.3)

plt.show();
```

## Task 4)

```python
import pandas as pd
plt.style.use('fivethirtyeight')  # Special plot style
```

```python
# Credit card transaction frauds database
cards = pd.read_csv('creditcard.csv')
cards.head(3)
```

```python
# Class 1 is specified as 'Fraud'
classes = cards.loc[:,'Class']
classes.value_counts()
```

```python
# Barplot of transaction classes
plt.figure(figsize = (4,6))
classes.value_counts().plot(kind='bar')

plt.show()
```

```python
# Barplot of transaction classes (log scaling)
plt.figure(figsize = (4,6))
classes.value_counts().plot(kind='bar')

plt.yscale('log')

plt.show()
```

```python
# Histograms for transaction feature 'v1' by classes

v1_clear = cards[cards['Class']==0]['V1']
v1_fraud = cards[cards['Class']==1]['V1']

plt.figure(figsize=(8,5))

plt.hist(v1_clear, bins=20, density=True, alpha=0.5, color='gray', label='Class 0')
plt.hist(v1_fraud, bins=20, density=True, alpha=0.5, color='red', label='Class 1 (frauds)')

plt.title('Feature "V1"', fontsize=14)
plt.ylabel('Density', fontsize=12)

plt.legend(loc="upper left")

plt.xlim(-35,5)

plt.show()
```
